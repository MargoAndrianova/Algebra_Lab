## Завдання 1. 

__Знайдіть усі незвідні системи твірних групи $S_4$.__

```python
S4 = SymmetricGroup(4)

print(f"Група: {S4}")

gener = S4.gens()
print("Твірні групи: ")
for gen in gener:
    print(gen)
```

## Завдання 2. 

__Знайдіть усі підгрупи груп $S_4, A_4, S_5, A_5$.__

```python
def list_subgroups(group):
    print(f"Підгрупи {group}:")
    subgroup = group.subgroups()
    for i, H in enumerate(subgroup):
        print(f"{i + 1}: Порядок: {H.order()}, Елементи: {H.list()}")
    print(f"Кількість: {len(subgroup)}\n")

S4 = SymmetricGroup(4)
A4 = AlternatingGroup(4)
S5 = SymmetricGroup(5)
A5 = AlternatingGroup(5)

list_subgroups(S4)
list_subgroups(A4)
list_subgroups(S5)
list_subgroups(A5)
```

## Завдання 3. 

__Задайте дві випадкові підстановки з групи $S_n, \, (1 \le n \le 100)$. Перевірте чи є вони парними. Знайдіть порядок групи, ними породженої. Чи буде ця група дорівнювати $A_n$? $S_n$?__

```python
to_rand = [i + 1 for i in range(100)]
n = choice(to_rand)

Sn = SymmetricGroup(n)
An = AlternatingGroup(n)

sigma = Sn.random_element()
tau = Sn.random_element()

print("Підстановка sigma:", sigma)
print("Підстановка tau:", tau)

sigma_is_even = sigma.sign() == 1
tau_is_even = tau.sign() == 1

print("sigma парна?", sigma_is_even)
print("tau парна?", tau_is_even)

# Створюємо групу, породжену sigma та tau
G = PermutationGroup([sigma, tau])

order_G = G.order()
print("Порядок групи:", order_G)

# Перевіряємо ізоморфізм з A_n та S_n
is_G_equal_An = G.is_isomorphic(An)
is_G_equal_Sn = G.is_isomorphic(Sn)

print("Група G ізоморфна A_n?", is_G_equal_An)
print("Група G ізоморфна S_n?", is_G_equal_Sn)
```
