## Завдання 1

Діедральна група $D_{20}$ має ряд нормальних підгруп: 

```python
G = DihedralGroup(20)

subgroups = G.normal_subgroups()
orders = [N.order() for N in subgroups]
print("Порядки нормальних підгруп:", orders)

for N in subgroups:
    phi = N
    print(f"Порядок підгрупи N: {N.order()}, Ядро: {phi.order()}")
    
    if phi == N:
        print("Ядро рівне підгрупі N")
    else:
        print("Ядро не рівне підгрупі N")
```

Для кожної нормальної підгрупи $N$ побудуйте ендоморфізм $\phi : D_{20} \rightarrow D_{20}$ такий, що $Ker \phi = N$. Задайте його через `PermutationGroupMorphism` і використайте метод `.kernel` як підтвердження. 

___Зауваження___: ядро має бути не __ізоморфне__ підгрупі, а __рівне__ їй.

## Завдання 2

Знайдіть центр та представників класів спряженості для груп $C_{11}, D_{13}, Q_{8}, SL(3, \mathbb{Z}_7)$. Виведіть кількість елементів в кожній групі, кількість класів спряженості та запишіть формулу для них. 

```python
# Циклічна група C11
C11 = CyclicPermutationGroup(11)
print("Циклічна група C11:")
print("Порядок групи:", C11.order())
print("Центр групи:", C11.center())
print("Кількість класів спряженості:", len(C11.conjugacy_classes_subgroups()))
print()

# Діедральна група D13
D13 = DihedralGroup(13)
print("Діедральна група D13:")
print("Порядок групи:", D13.order())
print("Центр групи:", D13.center())
print("Кількість класів спряженості:", len(D13.conjugacy_classes_subgroups()))
print()

# Кватерніонна група Q8
Q8 = QuaternionGroup()
print("Кватерніонна група Q8:")
print("Порядок групи:", Q8.order())
print("Центр групи:", Q8.center())
print("Кількість класів спряженості:", len(Q8.conjugacy_classes_subgroups()))
print()

# Спеціальна лінійна група SL(3, Z7)
SL3_Z7 = SL(3, GF(7))
print("Група SL(3, Z7):")
print("Порядок групи:", SL3_Z7.order())
print("Центр групи:", SL3_Z7.center())
print("Кількість класів спряженості:", len(SL3_Z7.conjugacy_classes()))
print()
```

1. Побудуйте граф гіперкуба $C_4$ і знайдіть його групу автоморфізмів. Розгляньте цю групу і її дію на граф. Покажіть орбіти цієї групової дії та зробіть висновки. 
2. Знайдіть стабілізатор довільної вершини та розгляньте його як групу з дією на весь граф. Знайдіть орбіти цієї нової групової дії. Зробіть висновки в термінах гіперкуба.

```python
G = graphs.CubeGraph(4)
G.show()

vertices = G.vertices()
print("Вершини графа:", vertices)

AutG = G.automorphism_group()
print("Група автоморфізмів графа:", AutG)

orbits_autG = AutG.orbits()
print("Орбіти групи автоморфізмів на вершинах графа:", orbits_autG)

vertex = vertices[0]

Stab_vertex = AutG.stabilizer(vertex)
print(f"Стабілізатор вершини {vertex}:", Stab_vertex)

orbits_stab = Stab_vertex.orbits()
print("Орбіти стабілізатора на вершинах графа:", orbits_stab)
```

Побудуйте простий граф, в якого група автоморфізмів тривіальна. 

```python
G = graphs.PathGraph(1)

G.show()

AutG = G.automorphism_group()
print("Група автоморфізмів графа:", AutG)
print("Порядок групи автоморфізмів:", AutG.order())

print("Чи є група автоморфізмів тривіальною?", AutG.order() == 1)
```
